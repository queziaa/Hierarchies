```python
#  feijiangtxt nplist

import os

os.chdir('feijiang/')
files = os.listdir()
files = [file for file in files if file.endswith('.txt')]
print(files)

import Hlib
for i in range(len(files)):
    with open(files[i], 'r') as f:
        temp = []
        lines = f.readlines()
        for line in lines:
            temp.append(line.split())
                    # float
        if len(temp) != 36:
            print('error')
        for i in range(len(temp)):
            if len(temp[i]) != 36:
                print('error')
            for j in range(len(temp[i])):
                temp[i][j] = float(temp[i][j])

    out,nonSample,trSample = Hlib.plt_seaborn(temp,6)

    ordered_positions = Hlib.find_ordered_positions(nonSample, trSample)

    print(files[i],ordered_positions)
    



    
```
